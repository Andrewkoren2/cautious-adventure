The double pendulum is one of the most common examples of how a slight change to an analytical, well-understood physical system (like a single pendulum) becomes chaotic with a very minor change.

The double pendulum is simply two pendulums stuck together



An illustration of a double pendulum (wikipedia)

## Equations of motion

Interestingly enough, the 

In order to numerically simulate the double pendulum, we'll find the Lagrangian and use that to create the equations of motion

$$L = T - U$$

In this case both linear and angular kinetic energy are present, but potential will just be gravitational $U(y)=mgy$

This produces

$$
L = \frac{m}{2}(v_1^2 + v_2^2) + \frac{I}{2}(\dot \theta_1^2 + \dot \theta_2^2) - mg(y_1+y_2) \\
= \frac{m}{2}(\dot x_1^2 + \dot y_1^2+ \dot x_2^2 + \dot y_2 ^2) + \frac{I}{2}(\dot \theta_1^2 + \dot \theta_2^2) - mg(y_1+y_2)
$$

Just as a pendulum is parameterized entirely by it's angle, the position of a double pendulum can be described using the angle each pendulum makes off the vertical axis. We'll find the center of mass as a function of angle $\theta$ off the vertical.

To simplify calculations, we'll first use the case where each pendulum has equal length and equal (constant) mass distributions. That means the pendulum bars themselves carry the mass, rather than a mass stuck to the end.

The center of mass for the first pendulum is the same as the usual pendulum, at half the length $l$ of the pendulum.

$$
cm_{x1}(\theta_1) = \frac{l}{2}\sin \theta_1 \\
cm_{y1}(\theta_1) = -\frac{l}{2}\cos \theta_1 
$$
The minus sign is used to make $cm_1(0) = \begin{pmatrix} 0 \\ -\frac{l}{2}\end{pmatrix}$

The second center of mass is located midway down the second pendulum. We can use the same equation if we include the displacement of the first pendulum

$$
cm_{x2}(\theta_1, \theta_2) = l\sin \theta_1 + \frac{l}{2}\sin \theta_2\\
cm_{y2}(\theta_1, \theta_2) -l\cos \theta_1 -\frac{l}{2}\cos \theta_2 
$$

Our lagrangian depends on the velocity and height of these centers of mass.

We'll take the derivative of each with respect to time

$$
\dot x_1 = \frac{l}{2} \dot \theta_1 \cos \theta_1 \implies \dot x_1^2 = \frac{l^2}{4} \dot \theta_1^2 \cos^2 \theta_1\\
\dot y_1 = \frac{l}{2} \dot \theta_1 \sin \theta_1 \implies \dot y_1^2 = \frac{l^2}{4} \dot \theta_1^2 \sin^2 \theta_1\\
\dot x_1^2 + \dot y_1^2 = \frac{l^2}{4} \dot \theta_1^2 (\sin^2 \theta_1 + \cos^2\theta_1) = \frac{l^2}{4} \dot \theta_1^2
$$

$$
\dot x_2 = 2 \dot x_1 + \frac{l}{2} \dot \theta_2 \cos \theta_2 \implies \dot x_2 = 4 \dot x_1 ^2 + 2l \dot x_1 \dot \theta_2 \cos \theta_2 + \frac{l^2}{4}\dot \theta_2 ^2 \cos ^2 \theta_2 \\
\dot y_2 = 2\dot y_1 + + \frac{l}{2} \dot \theta_2 \cos \theta_2 \implies \dot y_2^2 = 4 \dot y_1 ^2 + 2l \dot y_1 \dot \theta_2 \sin \theta_2 + \frac{l^2}{4}\dot \theta_2 ^2 \sin ^2 \theta_2\\

\dot x_2^2 + \dot y_2^2 = 4(\dot x_1^2 + \dot y_1^2) +  \left(\frac{l^2}{4} \dot \theta_2^2 (\cos^2 \theta_2 + \sin^2 \theta_2) + l^2\dot \theta_1 \dot \theta_2 (\cos \theta_1 \cos \theta_2 + \sin \theta _1 \sin \theta_2)\right) \\

= l^2 \dot \theta_1^2 + \frac{1}{4} \dot \theta_2^2 + \dot \theta_1 \dot\theta_2 \cos(\theta_1 - \theta_2)
$$

This gets us our linear kinetic energy and height as functions of angle and angular velocity. The moment of inertia for a pendulum swinging from its hinge is $I = \frac{1}{12}ml^2$, so rotational kinetic energy is already laid out. This yields the lagrangian

$$
L = \frac{m}{2} \left(\frac{5}{4}l^2 \dot \theta_1^2 + \frac{1}{4} \dot \theta_2^2 + \dot \theta_1 \dot\theta_2 \cos(\theta_1 - \theta_2)  \right) + \frac{ml^2}{24} \left(\dot \theta_1^2 + \dot \theta_2^2 \right) + \frac{m}{2} gl (3 \cos \theta_1 + \cos \theta_2) \\
= \frac{m}{6}l^2 \left(4\dot\theta_1^2 + \dot \theta_2^2 + 3 \dot \theta_1 \dot \theta_2 \cos(\theta_1 - \theta_2) \right) + \frac{m}{2} gl(3\cos \theta_1 + \cos \theta_2)
$$


Now we'll use this to get our equations of motion.

$$
\frac{dL}{d\theta_1} = -\frac{m}{2}l^2 \dot \theta_1 \dot \theta_2 \sin(\theta_1 -\theta_2) - \frac{3m}{2} gl \sin(\theta_1) \\
\frac{d}{dt}\frac{dL}{d\dot \theta_1} = \frac{d}{dt} \left(\frac{m}{6}l^2(3 \dot \theta_2 \cos(\theta_1 - \theta_2)) + ml^2 \frac{4}{3}\dot \theta_1 \right) = \frac{m}{6}l^2\left(3 \ddot \theta_2 \cos(\theta_1-\theta_2) - 3 \dot \theta_2(\dot \theta_1 - \dot \theta_2)\sin(\theta_1 - \theta_2) \right) + \frac{4ml^2}{3} \ddot \theta_1
$$

This yields our euler-lagrange
$$
\frac{d}{dt}\frac{dL}{d\dot \theta_1} = \frac{dL}{d\theta_1}
$$

$$
\frac{m}{6}l^2\left(3 \ddot \theta_2 \cos(\theta_1-\theta_2) - 3 \dot \theta_2(\dot \theta_1 - \dot \theta_2)\sin(\theta_1 - \theta_2) \right) + \frac{4ml^2}{3} \ddot \theta_1 = -\frac{m}{2}l^2 \dot \theta_1 \dot \theta_2 \sin(\theta_1 -\theta_2) - \frac{3m}{2} gl \sin(\theta_1) \\

\frac{4l^2}{3} \ddot \theta_1 = -\frac{1}{2}l^2 \dot \theta_1 \dot \theta_2 \sin(\theta_1 -\theta_2) - \frac{3}{2} gl \sin(\theta_1) - \frac{1}{6}l^2\left(3 \ddot \theta_2 \cos(\theta_1-\theta_2) - 3 \dot \theta_2(\dot \theta_1 - \dot \theta_2)\sin(\theta_1 - \theta_2) \right) \\


\ddot \theta_1 = -\frac{3}{8} \dot \theta_1 \dot \theta_2 \sin(\theta_1 -\theta_2) - \frac{9}{4} \frac{g}{l} \sin(\theta_1) - \frac{1}{8}\left(3 \ddot \theta_2 \cos(\theta_1-\theta_2) - 3 \dot \theta_2(\dot \theta_1 - \dot \theta_2)\sin(\theta_1 - \theta_2) \right)
$$
Which is actually dependent on $\ddot \theta_2$. This is a coupled system of differential equations which isn't easily solved.

Using some linear algebra tricks one can rewrite the equation of motion for both $\ddot \theta_1$ and $\ddot \theta_2$, but I'll leave that to the experts. Diego Assencio has a post on how this can be done on [his website](https://dassencio.org/33), although he uses point masses rather than using the pendulum arms as masses.

## Numerical Simulation

Since $\ddot \theta_1$ is related to $\ddot \theta_2$, we'll be using the Runge-Kutta method to simulate this motion. I'll be using a modified version of [dassencio](https://github.com/dassencio/double-pendulum)'s implementation into python to do this, which uses bobs attached at the end of the pendulums rather than calculating a bunch of centers of mass. 

```python
#!/usr/bin/env python3

"""Double pendulum simulator based on its Euler-Lagrange equations."""

import math
import numpy


class DoublePendulumLagrangian:

    def __init__(self, g, m1, m2, t1, t2, w1, w2, L1, L2):
        """
        Constructs a double pendulum simulator based on its
        Euler-Lagrange equations. Bob #1 is the one attached to the
        fixed pivot.

        g - The gravitational acceleration.
        m1 - The mass of bob #1.
        m2 - The mass of bob #2.
        t1 - The initial angle of bob #1.
        t2 - The initial angle of bob #2.
        w1 - The initial angular velocity of bob #1.
        w2 - The initial angular velocity of bob #2.
        L1 - The length of the rod for bob #1.
        L2 - The length of the rod for bob #2.
        """

        self.g = g
        self.m1 = m1
        self.m2 = m2
        self.t1 = t1
        self.t2 = t2
        self.w1 = w1
        self.w2 = w2
        self.L1 = L1
        self.L2 = L2

    def potential_energy(self):
        """Computes the potential energy of the system."""

        m1 = self.m1
        t1 = self.t1
        L1 = self.L1
        m2 = self.m2
        t2 = self.t2
        L2 = self.L2

        g = self.g

        # compute the height of each bob
        y1 = -L1 * math.cos(t1)
        y2 = y1 - L2 * math.cos(t2)

        return m1 * g * y1 + m2 * g * y2

    def kinetic_energy(self):
        """Computes the kinetic energy of the system."""

        m1 = self.m1
        t1 = self.t1
        w1 = self.w1
        L1 = self.L1
        m2 = self.m2
        t2 = self.t2
        w2 = self.w2
        L2 = self.L2

        # compute the kinetic energy of each bob
        K1 = 0.5 * m1 * (L1 * w1)**2
        K2 = 0.5 * m2 * ((L1 * w1)**2 + (L2 * w2)**2 +
                         2 * L1 * L2 * w1 * w2 * math.cos(t1 - t2))

        return K1 + K2

    def mechanical_energy(self):
        """
        Computes the mechanical energy (total energy) of the
        system.
        """

        return self.kinetic_energy() + self.potential_energy()

    def lagrange_rhs(self, t1, t2, w1, w2):
        """
        Computes the right-hand side of the Euler-Lagrange equations
        for the double pendulum and returns it as an array.

        t1 - The angle of bob #1.
        t2 - The angle of bob #2.
        w1 - The angular velocity of bob #1.
        w2 - The angular velocity of bob #2.
        """

        m1 = self.m1
        L1 = self.L1
        m2 = self.m2
        L2 = self.L2

        g = self.g

        a1 = (L2 / L1) * (m2 / (m1 + m2)) * math.cos(t1 - t2)
        a2 = (L1 / L2) * math.cos(t1 - t2)

        f1 = -(L2 / L1) * (m2 / (m1 + m2)) * (w2**2) * math.sin(t1 - t2) - \
            (g / L1) * math.sin(t1)
        f2 = (L1 / L2) * (w1**2) * math.sin(t1 - t2) - (g / L2) * math.sin(t2)

        g1 = (f1 - a1 * f2) / (1 - a1 * a2)
        g2 = (f2 - a2 * f1) / (1 - a1 * a2)

        return numpy.array([w1, w2, g1, g2])

    def time_step(self, dt):
        """
        Advances one time step using RK4 (classical Runge-Kutta
        method).
        """

        m1 = self.m1
        t1 = self.t1
        w1 = self.w1
        L1 = self.L1
        m2 = self.m2
        t2 = self.t2
        w2 = self.w2
        L2 = self.L2

        # y is an array with the generalized coordinates (angles +
        # angular velocities)
        y = numpy.array([t1, t2, w1, w2])

        # compute the RK4 constants
        k1 = self.lagrange_rhs(*y)
        k2 = self.lagrange_rhs(*(y + dt * k1 / 2))
        k3 = self.lagrange_rhs(*(y + dt * k2 / 2))
        k4 = self.lagrange_rhs(*(y + dt * k3))

        # compute the RK4 right-hand side
        R = 1.0 / 6.0 * dt * (k1 + 2.0 * k2 + 2.0 * k3 + k4)

        # update the angles and angular velocities
        self.t1 += R[0]
        self.t2 += R[1]
        self.w1 += R[2]
        self.w2 += R[3]
```

```python
import numpy as np

def end_location(angle1,angle2,L1,L2):
    x1 = L1 * np.sin(angle1)
    y1 = -L1 * np.cos(angle1)
    x2 = x1 + L2 * np.sin(angle2)
    y2 = y1 - L2 * np.cos(angle2)
    return [x1, y1, x2, y2]
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import PillowWriter

g = 9.81
m1 = 1
m2 = m1
L1 = 1
L2 = L1

angle1, angle2 = 0, 0
velocity1, velocity2 = 10, -10

dt = 0.01
time = 15
frames = time / dt
print('frames: ', frames)

p2 = DoublePendulumLagrangian(g, m1, m2, angle1, angle2, velocity1, velocity2, L1, L2)

p2_record = np.array(end_location(angle1, angle2, L1, L2))

metadata = {'title': 'Movie', 'artist': 'AndrewKoren'}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # Set the figure size here

capture_rate = 2 # captures 1/rate frames

writer = PillowWriter(fps=capture_rate/dt)

data = []

with writer.saving(fig, '2Pendulum.gif', 100):
    for tval in range(int(frames)):
        p2.time_step(dt)
        angle1, angle2 = p2.t1, p2.t2
        p2_record = np.vstack([p2_record, end_location(angle1, angle2, L1, L2)])

        if tval % capture_rate == 0:
            ax1.set_xlim(-L1 - L2, L1 + L2)
            ax1.set_ylim(-L1 - L2, L1 + L2)
            
            print(tval*dt)
            ax1.plot(p2_record[:,2][:tval+1], p2_record[:,3][:tval+1], color = 'r')
            ax1.plot((0,p2_record[tval][0]),(0,p2_record[tval][1]), color = 'g')
            ax1.plot((p2_record[tval][0],p2_record[tval][2]),(p2_record[tval][1],p2_record[tval][3]))

            ax1.set_title(f't = {tval*dt:.2f}')

            ax2.set_ylim(-40,40)
            ax2.bar(1, p2.kinetic_energy(), color='blue', label='Kinetic Energy')
            ax2.bar(2, p2.mechanical_energy(), color='green', label='Mechanical Energy')
            ax2.bar(3, p2.potential_energy(), color='red', label='Potential Energy')
            ax2.set_xticks([1, 2, 3])
            ax2.set_xticklabels(['Kinetic', 'Mechanical', 'Potential'])
            ax2.legend()


            writer.grab_frame()
            ax1.clear()
            ax2.clear()
```

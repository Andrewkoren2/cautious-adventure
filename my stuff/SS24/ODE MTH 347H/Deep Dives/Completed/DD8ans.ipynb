# Deep Dive #8 - Integrating Factor Method

Remember, matricies aren't actually scary. They just say that the rate of change of a variable is dependent on the current value of other variables. 

The integrating factor method introduced in section 1.4 for constant coefficients is $y'=ay \implies y= ce^{at}$ which, with intial value $y_0=y(0)$ this becomes $y=y_0e^{at}$

How do we do this when A is a matrix? We'll use the property of matrix exponentials 
<mark> 
$\left(e^{At}\right)'=e^{At}A $ 
</mark> as follows

$$
x'=Ax \\
x'-Ax={0} \\
e^{-At}x'-e^{-At}Ax=e^{-At}{0}\\
e^{-At}x'+e^{-At}(-A)x={0}\\
e^{-At}x'+\left(e^{-At}\right)'x={0}\\
\left(e^{-At}x\right)'={0} \\
e^{-At}x={c} \\
x(t)=e^{At}c \\
x(0)=e^0c=x^0\implies c=x^0\\
x(t)=e^{At}x^0
$$

$$

Can we go and integrate $e^{-At}{ b}$? Since $\left(e^{-At}\right)'=-A e^{-At}$ and ${ b}$ is constant, we have $$\left(-A^{-1}e^{-At} { b}\right)'=A^{-1}Ae^{-At}{ b} = e^{-At}{ b}\\ \implies \int e^{-At} { b }dt =-A^{-1}e^{-At} { b }+ { c  }$$

$$
\left(e^{-At}x \right)'= e^{-At}{ b }\implies e^{-At}x = -A^{-1}e^{-At} { b }+ { c }\\
x = e^{At}{ c }- A^{-1}e^{-At} { b} \\
x(0) = x^0 = c - A^{-1} { b} \implies c = x^0 + A^{-1}b \\
x(t) = e^{At}\left(x^0+A^{-1}{ b}\right) - A^{-1} { b} \\
$$

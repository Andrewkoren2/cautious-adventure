```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def a(t):
    return 3/t

def b(t):
    return 0

# Define the differential equation
def yprime(y, t, a, b):
    at = a(t)
    bt = b(t)
    dydt = at * y * y + bt
    return dydt

# Initial condition
y0 = 1

# Time points
t = np.linspace(1, 11, 100)

# Solve ODE using lambda to pass the current a(t) and b(t)
y = odeint(lambda y, t: yprime(y, t, a, b), y0, t)

plt.plot(t,y)
```

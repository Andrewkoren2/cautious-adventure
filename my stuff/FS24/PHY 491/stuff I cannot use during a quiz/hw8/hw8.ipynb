Before we get started on this problem, let's have a look at what this potential looks like for the square lattice.

```python
import numpy as np
from numpy import cos, pi

import matplotlib.pyplot as plt

e0 = 1
a = 1
params = [e0, a]

x = np.linspace(-pi/a, pi/a, 100)
y = np.linspace(-pi/a, pi/a, 100)

kx, ky = np.meshgrid(x, y)

def tight_binding(kx, ky, params):
    e0, a = params
    return e0*(2-cos(a*kx)-cos(a*ky))

energy = tight_binding(kx, ky, params)

ax = plt.figure().add_subplot(111, projection='3d')
ax.plot_surface(kx, ky, energy, cmap='viridis')
ax.view_init(35, 20)
plt.title('Energy, arbitrary units')
plt.tight_layout()
```

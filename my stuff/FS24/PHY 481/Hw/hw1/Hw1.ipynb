Part 1: problem 1.4

The vector perpenicular to a plane is the cross product of two vectors on the surface of the plane. Given the verticies of the triangle in the figure, we can create vectors on the surface

$$
\vec a = (1, 0, 0) \rightarrow (0, 0, 3) \\
\vec b = (1, 0, 0) \rightarrow (0, 2, 0)
$$

The separation vector for each is thus

$$
\vec a = \langle-1, 0, 3\rangle \\
\vec b = \langle-1, 2, 0\rangle \\
$$

The cross product is

$$
\vec a \times \vec b = \left| \begin{matrix} \hat x & \hat y & \hat z \\ -1 & 0 & 3 \\ -1 & 2 & 0\end{matrix} \right| = -2 \hat z -3\hat y
$$

Well, that clearly points in the wrong direction. Remember right hand rule...

$$
\vec b \times \vec a = 2 \hat z + 3\hat y
$$

A unit vector has magnitude 1, so 

$$
\hat n = \frac{2 \hat z + 3\hat y}{\left|2 \hat z + 3\hat y\right|} = \frac{2 \hat z + 3\hat y}{\sqrt{13}}
$$

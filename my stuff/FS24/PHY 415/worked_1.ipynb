1. Mechanical System of Study

We will look at a constrained equation of motion for a rollercoaster. Specifically, the one used by rollecoasters in real life, the loop with constant centripedal motion. Circular loops are avoided since the forces on riders very greatly as they move up and down the loop due to the combination of centripedal motion and constant downward force producing vastly varying velocities. 



A loop with constant centripedal motion
Image from Physics Education Journal <br>
Physics Education 40, p 517 (2005)

Centripedal acceleration as in a vertical loop comes from 

$$
a_c = \frac{2gh}{r}
$$

To keep this value constant, the following curve can be used

$$
\frac{1}{r} = \frac{1}{h}\frac{h_0}{r_0}
$$

Where $h_0$ and $r_0$ are the loop height and radius factor, and can be changed to describe various loops of constant centripedal motion. As such, the ratio between $r$ and $h$ yields a constant $a_{c,0}=\frac{2gh_0}{r_0}$. This result allows for engineers to control the maximum g-force a person experiences and prevent brain damage due to strong forces. 

Note: I originally tried a more complex rollercoaster loop, then realized how painful it would be.

```python
# Shape of the constraint equation
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, cos, pi

def polar_to_x(r, theta):
    return r * sin(theta)

def polar_to_y(r, theta):
    return -r * cos(theta)

def height(h0, theta):
    return h0 * (1-cos(theta))

h0 = 1
r0 = 10
g = 9.81

print(f'Centripedal acceleration: {2*g*h0/r0}')

xvals = []
yvals = []
for theta in np.linspace(0, 2*pi, 1000):
    h = height(h0, theta)
    r = h*r0/h0
    x,y = polar_to_x(r,theta), polar_to_y(r,theta)
    
    xvals.append(x)
    yvals.append(y)

plt.plot(xvals, yvals);
```

2. Developing the Lagrangian and Assumptions/Simplifications

Using polar coordinates is ideal since the curve depends on $r$ and rotational energy is important.

Using $\theta = 0$ as down and $h =h_0 (1- \cos(\theta))$, we can actually change our constraint equation to relate $r$ and $\theta$

$$
r = \frac{hr_0}{h_0} = \frac{h_0 (1- \cos\theta)r_0}{h_0} = r_0(1-\cos\theta)
$$

This also relates our velocities
$$
\dot r  = r_0 \sin \theta \dot \theta
$$
The rate of change of $r$ with respect to $\theta$ is greater near the middle. This is because the rollercoaster slows/speeds up the fastest when gravity is parallel to the cart, and the coaster's local radius must expand/contract to keep centripedal acceleration constant.

From here our potential is very simple
$$
U = mgh = mgh_0 (1- \cos(\theta)) \\
$$
And we can make our kinetic energy dependant on angular rotation as well
$$
T = KE_{\text{linear}} + KE_{\text{rot}} = \frac{1}{2}m \left( \dot r^2 + r^2 \dot \theta^2 \right) + \frac{1}{2} m r^2 \dot \theta^2 \\
= \frac{1}{2} m \left[\left( r_0 \sin \theta \dot \theta \right)^2 + 2r_0^2 (1-\cos\theta)^2 \dot \theta ^2 \right] \\
= \frac{1}{2} m r_0 ^2 \dot \theta^2 \left[\sin^2(\theta)  + 2(1-\cos(\theta))^2 \right] \\
= \frac{1}{2} m r_0 ^2 \dot \theta^2 \left[3-4\cos(\theta)+\cos^2(\theta) \right]
$$ 

Thus we have our Lagrangian with two coordinates:

$$
\mathcal{L} = T - U = \frac{1}{2} m r_0 ^2 \dot \theta^2 \left[3-4\cos(\theta)+\cos^2(\theta) \right] - mgh_0 (1- \cos(\theta))
$$

The largest assumption we made was that there is no friction on the system and thus no damping force. We also don't account for the track before/after the loop.

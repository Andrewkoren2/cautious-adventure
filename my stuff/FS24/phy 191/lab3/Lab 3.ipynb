Centripedal motion. The force of gravity is necessary to maintain circular orbit around the earth. The satellites are moving so fast that they 'miss' the earth as they fall

(a) see plot below

```python
import matplotlib.pyplot as plt
import numpy as np

tspan = np.linspace(0, 10, 200)

def y(t, g, v0, y0):
    return -g*t**2/2 + v0*t + y0

plt.plot(tspan, y(tspan, 9.81, 0, 1000))
plt.xlabel('time (s)')
plt.ylabel('height (m)');
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

avg_circumfrence = np.array([34.57, 35.45, 35.51, 35.45, 36.00])

avg_radius = avg_circumfrence/(2*np.pi)
avg_vol = 4/3*np.pi*avg_radius**3
avg_vol_m = avg_vol * 1e-6

tempc = np.array([-1, 10, 21, 30, 35])
tempk = tempc + 273.15

plt.scatter(tempk, avg_vol)
plt.xlabel('temperature (kelvn)')
plt.ylabel(f'volume (cm$^3$)')

def linear(x, m):
    return m*x #+ y0

params, _ = curve_fit(linear, xdata=tempk, ydata=avg_vol)

temp_fit = np.linspace(265, 325, 1000)
vol_fit = linear(temp_fit, *params)

plt.plot(temp_fit, vol_fit)

params[0]
```

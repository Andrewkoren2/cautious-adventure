```python
concentration = np.arange(0, 400+1, 20)


absorbance = [
1.91 ,1.81 ,1.72 ,1.62 ,1.53 ,1.43 ,1.34 ,1.24 ,1.14 ,1.05 ,.95 ,.86 ,.76 ,.67 ,.57 ,.48 ,.38 ,.29 ,.19 ,.10 ,0
]
absorbance.reverse()

plt.title('Concentration vs Absorbance')
plt.xlabel(r'concentration ($\mu M$)')
plt.ylabel('Absorbance')
plt.scatter(concentration, absorbance, label='collected data')

from scipy.optimize import curve_fit

def beers_law(x, m):
    return m*x

args, _ = curve_fit(beers_law, concentration, absorbance)

x = np.linspace(min(concentration), max(concentration), 1000)
plt.plot(x, beers_law(x, *args), color='red', label = f'y = {args[0]:.5f}x')
plt.legend()

```
